Exemple d'application de Kmeans
===============================

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd

df = pd.read_csv('donnees_kmeans.csv')
print("L'échantillon d'apprentissage")
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("Représentation des données")
plt.xlabel('X1')
plt.ylabel('X2')
plt.scatter(df['X1'], df['X2'], c='gray',s=50, alpha=0.5)

plt.grid()
plt.show()
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=2)
km.fit(df)
centroids = km.cluster_centers_
 
print('Les centres des clusters obtenus:')
print(centroids)
print(type(centroids))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("Représentation des donées à la fin du clustering")
plt.xlabel('X1')
plt.ylabel('X2')

plt.scatter(df['X1'], df['X2'], c= km.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1],  marker='x', c='black', s=50)
```

## Le même exemple avec annotation des points dans les figures

```python
import pandas as pd
df = pd.read_csv('donnees_kmeans.csv')
print("L'échantillon d'apprentissage")
print(df)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("Représentation des données")
plt.xlabel('X1')
plt.ylabel('X2')
plt.scatter(df['X1'], df['X2'], c='gray',s=50, alpha=0.5)

x = df['X1'].tolist() 
y = df['X2'].tolist() 

z = list(range(8))
for i in range(len(z)):
    plt.annotate(z[i],(x[i], y[i] + 0.2))


# adjusting the scale of the axes
plt.xlim((0, 8))
plt.ylim((0, 5))

plt.grid()
plt.show()  
    
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=2)
km.fit(df)
```

```python

centroids = km.cluster_centers_
inertia = km.inertia_

print('Les centres des clusters obtenus:')
print(centroids)

print("L'inertie du résultat est:", inertia)
```

```python

plt.title("Représentation des données à la fin du clustering")
plt.xlabel('X1')
plt.ylabel('X2')
plt.scatter(df['X1'], df['X2'], c= km.labels_.astype(float), s=50, alpha=0.5)

x = df['X1'].tolist() 
y = df['X2'].tolist() 
z = list(range(8))
for i in range(len(z)):
    plt.annotate(z[i],(x[i], y[i] + 0.2))

plt.scatter(centroids[:, 0], centroids[:, 1],  marker='x', c='black', s=30)    

# adjusting the scale of the axes
plt.xlim((0, 8))
plt.ylim((0, 5))

plt.grid()
plt.show()  
```

## Autre exemple

```python
data = {'X1': [25,34,22,27,33,33,31,22,35,34,67,54,57,43,50,57,59,52,65,47,49,48,35,33,44,45,38,43,51,46],
        'X2': [79,51,53,78,59,74,73,57,69,75,51,32,40,47,53,36,35,58,59,50,25,20,14,12,20,5,29,27,8,7]
       }
  
df = pd.DataFrame(data)
print("L'échantillon d'apprentissage")
print(df)
```

```python
import matplotlib.pyplot as plt
plt.title("Représentation graphique")
plt.xlabel('X1')
plt.ylabel('X2')
plt.scatter(df['X1'], df['X2'])
plt.grid()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3).fit(df)
centroids = km.cluster_centers_
inertia = km.inertia_

print('Les centres des clusters obtenus:')
print(centroids)

print("L'inertie du résultat est:", inertia)
plt.title("Représentation graphique à la fin du clustering")
plt.xlabel('X1')
plt.ylabel('X2')

plt.scatter(df['X1'], df['X2'], c= km.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', marker='x', s=50)
plt.grid()
```
