```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
df  = pd.read_csv("../boston.csv")
df

```

```python
x = df.drop("MEDV", axis = 1)
y = df["MEDV"]
print("Data sizes : \n" , "x:" ,x.shape , "\t y: " , y.shape)
```

```python
from sklearn.model_selection import train_test_split # type: ignore
x_train , x_test ,y_train ,y_test  = train_test_split(x,y , test_size=0.33 , random_state=123)
print("Train size :" , x_train.shape , y_train.shape)
print("Test size : ", x_test.shape , y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsRegressor # type: ignore
knn_reg_model = KNeighborsRegressor(n_neighbors = 15)
knn_reg_model.fit(x_train,y_train) 
```

```python
y_pred = knn_reg_model.predict(x_test)
print("prediction pour les exemples de teste :" )
df_temp = pd.DataFrame({"y_test :" : y_test  , "y_pred": y_pred})
print(y_pred)
```

```python
from sklearn import metrics
r2 = metrics.r2_score(y_test , y_pred)
mae = metrics.mean_absolute_error(y_test  , y_pred) 
mse = metrics.mean_squared_error(y_test , y_pred)
rmse = np.sqrt(mse)
print("r2 score :" , r2) 
print( "Mean Absolute error: " , mae)
print( "Mean Squared Error: " , mse )
print("Root mean Squared Error : " , rmse)
```
