## Trouver la valeur optimale de k avec la méthode du coude (elbow method)



```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd
df = pd.read_csv('donnees_kmeans_elbow.csv')
print("L'échantillon d'apprentissage")
df
```

```python
import matplotlib.pyplot as plt
plt.title("Représentation graphique")
plt.xlabel('X1')
plt.ylabel('X2')
plt.scatter(df['X1'], df['X2'])
plt.grid()
```

```python
# Trouver la valeur optimale de k avec la méthode du coude (elbow method)

from sklearn.cluster import KMeans

inertia_list = [] 
for k in range(1, 10): 
    km = KMeans(n_clusters = k)
    km.fit(df) 
    inertia_list.append(km.inertia_)

k_list = list(range(1, 10))

print('Valeurs de k:') 
print(k_list)
print('inerties correspondantes:') 
print(inertia_list)
```

```python
plt.plot(k_list, inertia_list)    
plt.title("Inertie en fonction de k")
plt.xlabel('k')
plt.ylabel('inertie')
plt.grid()
```

```python
plt.plot(k_list, inertia_list)  
plt.title("Inertie en fonction de k")
plt.xlabel('k')
plt.ylabel('inertie')
plt.axvline(3, linestyle='--', color='r')
plt.grid()
```

```python
print('Pour k = 3:')
print('-----------')

km = KMeans(n_clusters=3)
km.fit(df)

centroids = km.cluster_centers_
inertia = km.inertia_

print('Les centres des clusters obtenus:')
print(centroids)

print("L'inertie du résultat est:", inertia)

plt.title("Représentation graphique à la fin du clustering")
plt.xlabel('X1')
plt.ylabel('X2')
plt.scatter(df['X1'], df['X2'], c= km.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', marker = 'x', s=50)
plt.grid()
plt.show()
```
